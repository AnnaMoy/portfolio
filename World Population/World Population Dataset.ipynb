# Data Visualization

```python
import matplotlib.pyplot as plt

# group continent together and get the population
continent_df = df.groupby('Continent')[['2022 Population', '2020 Population']].sum()

#plot a bar chart
continent_df.plot(kind= 'bar')
# title the plot
plt.title('Total Population in 2020 and 2022 for each Continent')
#label x axis
plt.xlabel('Continent')
# label y axis
plt.ylabel('Years')
#changing the legend font size to 12 and position to upper left
plt.legend(loc= 'upper right', fontsize = 12)
#show plot
plt.show()
```

```python
# Plot data for Continent and the average population 
avg_population = df.groupby('Continent')['2022 Population'].mean()
print(avg_population)

#plot a line chart with continent and average population with markers,line color = green and linewidth
avg_population.plot(kind= 'line', marker = 'o', color = 'g', linewidth = 1)
# title the chart
plt.title('Average 2022 Population by Continent', fontsize = 15)
#label x axis
plt.xlabel('Continent', fontsize = 10)
# label y axis
plt.ylabel('2022 Population', fontsize = 10)
#changing the legend on the upper left side outside of the chart
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
# change yticks
plt.yticks([2e7, 4e7, 6e7, 8e7], ['20M', '40M', '60M', '80M'])
# add annotation
asia_index = avg_population.index.get_loc('Asia')
europe_index = avg_population.index.get_loc('Europe')
plt.annotate('90M in Population', 
             xy=(asia_index, avg_population['Asia']),  # Coordinates of Asia's data point
             xytext=(europe_index, 94_000_000),  # Adjust the text position
             arrowprops=dict(facecolor='blue', shrink=0.05))  # Add arrow pointing to the data point

#show plot
plt.show()
```

The average 2022 population was highest in Asia with an average population of 90 million. 

```python
import seaborn as sns

# Calculate the mean population by country/territory
avg_country = df.groupby('Country/Territory')['2022 Population'].mean().reset_index()

# Sort by population and select top 5 countries for clarity
top5_country = avg_country.sort_values(by='2022 Population', ascending=False).head(5)


# Create a bar plot for average 2022 Population by Country/Terrority and getting the mean for salary 
sns.barplot(data=top5_country, x='Country/Territory', y='2022 Population', estimator='mean', palette='muted')

# Title and labels
plt.title('Average 2022 Population by Country/Terrority')
plt.xlabel('Country/Terrority')
plt.ylabel('Average 2022 Population')
# Rotate x-axis labels
plt.xticks(rotation=45)
# rename the yticks
plt.yticks([.2e9, .4e9, .6e9, .8e9, 1e9, 1.2e9, 1.4e9], ['.2B','.4B', '.6B','.8B', '1.0B', '1.2B', '1.4B'])
# Show the plot
plt.show()
```

The top 5 average population for 2022 is the highest in China then India.
